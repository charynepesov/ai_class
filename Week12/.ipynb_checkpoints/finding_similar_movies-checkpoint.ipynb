```python
import pandas as pd
```

```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))
print(ratings.head())

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', encoding='iso-8859-1', names=m_cols, usecols=range(2))
print(movies.head())
```

```python
ratings = pd.merge(movies, ratings)
#print(ratings)
ratings.head()
```

```python
ratings.title.nunique(), ratings.user_id.nunique()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'], columns=['title'], values='rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head(20)
```

```python
starWarsRatings.value_counts(dropna=False)
```

```python
movieRatings[['101 Dalmatians (1996)', 'Star Wars (1977)']].corr()
```

```python
starWarsRatings.head()
```

```python
movieRatings.corrwith(starWarsRatings)
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
similarMovies
```

```python
similarMovies.sort_values(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head(10)
```

**Some movies have been rated by only a few people so let@s filter the result for movies which have greater than 100 rewiews**

```python
popularMovies = movieStats[movieStats['rating']['size'] >= 100]
```

```python
popularMovies.sort_values(('rating', 'mean'), ascending=False)
```
