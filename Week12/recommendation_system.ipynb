```python
r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols, usecols=range(3))
print(ratings.head())

m_cols = ['movie_id', 'title']
movies = pd.read_csv('ml-100k/u.item', sep='|', encoding='iso-8859-1', names=m_cols, usecols=range(2))
print(movies.head())
```

```python
ratings = pd.merge(movies, ratings)
#print(ratings)
```

```python
ratings.head()
```

```python
ratings.title.nunique(), ratings.user_id.nunique()
```

```python
movies_grouped = ratings.groupby('title').agg({'rating': [np.size, np.sum, np.mean]})
```

```python
movies_grouped
```

```python
popular_movies = movies_grouped.sort_values(('rating', 'mean'), ascending=False)
```

```python
popular_movies.head()
```

```python
grouped_sum = movies_grouped['rating']['sum'].sum()
grouped_sum
```

```python
popular_movies['percentage'] = popular_movies['rating', 'sum'].div(grouped_sum) * 100
```

```python
popular_movies.head()
```

```python
popular_movies.sort_values(('percentage'), ascending=False)
```

```python
popular_movies['Rank'] = popular_movies['percentage'].rank(ascending=False, method='min')
```
