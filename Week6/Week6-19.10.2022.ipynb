```python
def features_cleaner(data):
    clean_txt = []
    for i in range(0, len(data)):
        text=re.sub('[^a-zA-Z]', ' ', data[i])
        text=text.lower()
        text=text.split()
        ps=PorterStemmer()
        text=[ps.stem(j) for j in text if not j in set(stopwords.words('english'))]
        text=''.join(text)
        clean_txt.append(text)
    return clean_txt
```
