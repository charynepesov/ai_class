```python
from gtts import gTTS
myText = "Merhaba öğrenciler, yapay zeka dersimize hoş geldiniz..."

output = gTTS(text=myText, lang="ru", slow=False)
output.save('outputtr.mp3')
```

```python
fh = open('hello_world.txt', 'r')
myText = fh.read().replace("\n", " ")

output = gTTS(text=myText, lang="ru", slow=False)

output.save('hello_world.mp3')
fh.close()
```
